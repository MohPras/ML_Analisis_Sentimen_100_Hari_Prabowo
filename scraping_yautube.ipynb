# Scraping data analisis sentimen #100HariPemerintahanPrabowoGibran

```python
import googleapiclient.discovery
import googleapiclient.errors
import csv

# Set API key YouTube v3 kamu
api_key = 'AIzaSyCFJK_XpbTVaXF60ztqIX0E9VyU-ypwiPc'

# Daftar ID video yang ingin diambil komentarnya
video_ids = [
    'FZdQNw1AhZM',
    '814oQRHuunA',
    'xJ4S7o31nWI',
    'd9d4PAO-H5w',
    'XqB0YqqRYf8',
    'W8UJd_05poE'
]

# Fungsi untuk membuat service YouTube API
def get_youtube_service():
    youtube = googleapiclient.discovery.build(
        "youtube", "v3", developerKey=api_key
    )
    return youtube

# Fungsi untuk mengambil komentar dari video
def get_comments(youtube, video_id):
    comments = []
    request = youtube.commentThreads().list(
        part="snippet,replies",
        videoId=video_id,
        textFormat="plainText",
        maxResults=100  # maksimal 100 komentar per permintaan
    )
    
    while request:
        response = request.execute()
        
        # Menyaring dan menyimpan data komentar yang diambil
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comment_data = {
                'video_id': video_id,
                'username': comment['authorDisplayName'],
                'comment': comment['textDisplay'],
                'like_count': comment['likeCount'],
                'published_at': comment['publishedAt'],
                'reply_count': item['snippet']['totalReplyCount']
            }
            comments.append(comment_data)

        # Melanjutkan ke komentar berikutnya (jika ada)
        request = youtube.commentThreads().list_next(request, response)
    
    return comments

# Menyimpan komentar ke file CSV
def save_comments_to_csv(comments, filename='100-Hari-Prabowo-Gibran.csv'):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['video_id', 'username', 'comment', 'like_count', 'published_at', 'reply_count'])
        writer.writeheader()
        writer.writerows(comments)
    print(f"Comments saved to {filename}")

def main():
    youtube = get_youtube_service()
    all_comments = []

    for video_id in video_ids:
        comments = get_comments(youtube, video_id)
        all_comments.extend(comments)
    
    save_comments_to_csv(all_comments)

if __name__ == "__main__":
    main()
```
